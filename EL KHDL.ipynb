# E-learning: Nhập 2 số cuối số sinh viên, **chia 5 lấy phần dư**, tùy kết quả sẽ trả về giá trị khác nhau thông qua `dict`

- Các biến được sử dụng:  
`x` : là **2 số cuối của mã số sinh viên** được nhập từ bên ngoài vào chương trình  
`y` : được khởi tạo để **chia 5 lấy phần dư** của `x`
- Các class được sử dụng:  
`bangGiaTri` : là `dict` với các key là kết quả của `y` và value tuân theo 1 quy luật tương ứng **0:110 ; 1:29 ; 2:38 ; 3:47 ; 4:56**

```python
#ĐỀ: Nhập 2 số cuối của số sinh viên và chia cho 5, dư bao nhiêu sẽ trả lại giá trị khác nhau
#Nhập 2 số cuối của số sinh viên vào chương trình
x = int(input("Nhập 2 số cuối mã số sinh viên: "))

#Chia 5 lấy phần dư của số được nhập vào
y = x % 5 

#Bảng giá trị tương ứng với key của chúng sẽ có value riêng
bangGiaTri = {
    0: 110,
    1: 29,
    2: 38,
    3: 47,
    4: 56,
}

#In ra các giá trị tương ứng từ bangGiaTri 
print(bangGiaTri[y])
```

# EXERCISE 3: Viết chương trình xác định một số bất kì là số chẵn hay lẻ
- Các biến được sử dụng:  
`x` : là số cần được xác định, nhập từ bên ngoài vào chương trình  
- Các keyword được sử dụng:  
`if` : được sử dụng để kiểm tra **điều kiện**  
`else` : chạy nếu điều kiện `if` **sai**

```python
#Nhập số cần xác định vào chương trình
x = int(input("Nhập số tại đây: "))

#Nếu x chia hết cho 2 (dư = 0) thì chương trình sẽ in ra kết quả x là số chẵn
if x % 2 == 0:
    print(f"{x} là số chẵn")

#Ngược lại thì chương trình sẽ in ra kết quả x là số lẻ
else:
    print(f"{x} là số lẻ")
```

# EXERCISE 8: Viết chương trình phân biệt một số bất kì có phải là số nguyên tố hay không  
- Các biến được sử dụng:  
`x` : là số cần được xác định, được nhập vào chương trình từ bên ngoài  
`i` : có giá trị chạy từ 2 đến `sqrt(x)`  
- Các keyword được sử dụng:  
`if` : được sử dụng để **kiểm tra điều kiện** `x` có **chia hết** cho `i` hay không  
`else` : sẽ chạy khi điều kiện `if` **sai**  
`break` : **dừng ngay vòng lặp** `for` khi phát hiện `x % i == 0`  
`range` : tạo ra các giá trị của `i` chạy từ 2 đến `sqrt(x)` 
`for` : dùng để **lặp** qua các giá trị của `i` kiểm tra xem có giá trị nào của `i` chia hết cho `x`  

```python
#Truy cập vào thư viện để sử dụng căn bậc 2
import math

#Nhập dữ liệu cần xác định vào chương trình
x = int(input("Nhập số tại đây: "))

#Nếu x nhỏ hơn 2 thì chương trình sẽ in ra kết quả "x khôn phải số nguyên tố"
if x < 2:
    print(f"{x} không phải số nguyên tố")

#Nếu x lớn hơn 2, cho các giá trị của i chạy từ 2 đến căn x 
else:
    for i in range(2, int(math.sqrt(x)) + 1):
        
        #Nếu x chia hết cho i chứng tỏ x có nhiều hơn 2 ước đồng nghĩa với x không phải số nguyên tố
        #Lúc này chương trình sẽ in ra kết quả "x không phải số nguyên tố" và dừng vòng lặp for bằng break
        if x % i == 0:
            print(f"{x} không phải số nguyên tố")
            break
    
    #Trường hợp còn lại là số nguyên tố, chương trình sẽ in ra kết quả "x là số nguyên tố"
    else:
        print(f"{x} là số nguyên tố")
```
